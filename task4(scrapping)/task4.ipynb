```python
import requests
import random
import json

from bs4 import BeautifulSoup

HEADERS = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
BASE_URL = 'https://www.imdb.com'

def get_movies(url: str):
    response = requests.get(url, headers=HEADERS)
    soup = BeautifulSoup(response.text, "html.parser")
    movies = soup.find_all("div", class_="ipc-title")
    random.shuffle(movies)
    return movies

def get_base_movie_info(movie):
    movie_a_tag = movie.a
    movie_title = movie_a_tag.h3.contents[0]
    movie_url = movie_a_tag.attrs['href']
    return movie_title, movie_url

def get_full_movie_info(url: str):
    response = requests.get(f'{BASE_URL}{url}', headers=HEADERS)
    soup = BeautifulSoup(response.text, "html.parser")

    sub_url_index = url.rindex('?ref_')
    sub_url = url[:sub_url_index]
    year_element = soup.find('a', attrs={"href": f'{sub_url}releaseinfo?ref_=tt_ov_rdat'})
    if year_element:
        year = year_element.contents[0]
    else:
        year = "N/A"

    description_element = soup.find('p', attrs={"data-testid": "plot"})
    if description_element:
        description = description_element.contents[-1].contents[0]
    else:
        description = "N/A"

    return year, description


def pick_movies():
    count = 0
    print("----------------------------------")
    movies = get_movies(f'{BASE_URL}/chart/top')
    for movie in movies:
        movie_title, movie_url = get_base_movie_info(movie)
        year, description = get_full_movie_info(movie_url)
        print(f'{movie_title} ({year})')
        print(description)
        print("----------------------------------")
        count += 1
        if count == 10:
            break

if __name__ == "__main__":
    pick_movies()
```
